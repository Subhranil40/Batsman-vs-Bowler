# Function to get batsman vs bowler

```python
batsmen=delivery['batter'].unique().tolist()
batsmen.sort()
batsmen
```

```python
def batter_vs_bowler(batsman,bowler):
    m1=delivery['batter']==batsman
    m2=delivery['bowler']==bowler
    final=delivery[m1 & m2]
    # print("No. of Balls: \t Runs Scored \t Strike Rate \t No.of Dismissals \t No. of 6's \t No. of 4's \t No. of dot's ")
    # print(f"{final.shape[0]} \t {final['batsman_run'].sum()} \t {round(final['batsman_run'].sum()/final.shape[0],2)} \t {final.dropna(subset='player_out')[new.dropna(subset='player_out')['kind']!='run out'].shape[0]} \t {final[final['batsman_run']==6].shape[0]} \t {final[final['batsman_run']==4].shape[0]} \t {final[final['batsman_run']==0].shape[0]}")
    # Assuming 'final' is a DataFrame with the necessary data
    l1=['No. of Balls','Runs Scored','Strike Rate','No. of Dismissals',"No. of 6's","No. of 4's","No. of dot's"]
    l2=[final.shape[0],final['batsman_run'].sum(),round(final['batsman_run'].sum()/final.shape[0]*100,2),final.dropna(subset='player_out')[final.dropna(subset='player_out')['kind'] != 'run out'].shape[0],final[final['batsman_run'] == 6].shape[0],final[final['batsman_run'] == 4].shape[0],final[final['batsman_run'] == 0].shape[0]]
    df=pd.DataFrame([l2],columns=l1)
    return df
    
 
    
```

```python
batsman='V Kohli'
bowler='JJ Bumrah'
batter_vs_bowler(batsman,bowler)
```
